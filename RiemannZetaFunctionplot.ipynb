Attempt 1 Normal Zeta
$$\zeta(s)= \sum_{n=1}^{\infin} \frac{1}{n^s}$$

```python
import cmath

j = cmath.sqrt(-1)

def zeta_function(x,t):
    points=[]
    for x in range(1,x):
        for t in range(1,t):
            s=(1/2)+t*j
        result=(1/(x)**s)
        points.append([x,result])

    return points


    
```

```python
points=zeta_function(20,20)
x=[point[0] for point in points]
y=[point[1].real for point in points]
y_img=[point[1].imag for point in points]
plt.plot(x,y)
plt.grid()
plt.axvline(x=2, color='r', linestyle='--')
plt.axvline(x=3, color='r', linestyle='--')
plt.axvline(x=5, color='r', linestyle='--')
plt.axvline(x=7, color='r', linestyle='--')
plt.axvline(x=11, color='r', linestyle='--')
plt.axvline(x=13, color='r', linestyle='--')
plt.axvline(x=17, color='r', linestyle='--')
```
