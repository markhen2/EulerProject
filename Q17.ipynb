```python
units={0:'',1:'One',2:'two',3:'three',4:'four',5:'five',6:'sox',7:'seven',8:'eight',9:'nine'}

tens={10:'ten',20:'twenty',30:'thirty',40:'fourty',50:'fifty',60:'sixty',70:'seventy',80:'Eighty',90:'ninty'}

hundreds={100:'one hundred',200:'two hundred',300:'three hundred',400:'four hundred',500:'five hundred',600:'six hundred',700:'seven hundred',800:'eight hundred',900:'nine hundred'}

Specials={11:'eleven',12:'twelve',13:'Thirteen',14:'fourteen',15:'fifteen',16:'sixteen',17:'seventeen',18:'eighteen',19:'nineteen',1000:'One thousand'}
```

```python
def num(n):
    s_n=str(n)
    if n< 10:
        return units[n]
    elif n==10:
        return tens[10]    
    elif 10 < n <= 99:
           if n in Specials:
                return Specials[n]
           elif n not in Specials:
               
                ten=int(s_n[0])*10
                d_ten=tens[ten]
                unit=int(s_n[1])
                d_unit=units[unit]
                return d_ten + ' ' + d_unit
    elif 100<=n<998:
        
        hun=int(s_n[0])*100
        ten=int(s_n[1])*10
        unit=int(s_n[2])

        d_hun=hundreds[hun]

        special_check=ten+unit
     
        if special_check in Specials:
             d_ten=Specials[special_check]
             d_unit=''
             return d_hun+' and '+d_ten+d_unit 

        elif ten+unit==0: 
             return d_hun
        
        elif special_check not in Specials:
             if ten==0:
                  d_ten=''
             else:
               d_ten=tens[ten]
             if unit==0:
               d_unit=''
             else: 
                 d_unit=units[unit]
 
        
        return d_hun+' and '+d_ten+' '+d_unit
    

    elif n==998:
        return hundreds[900]+' and '+tens[90]+' '+units[8]

    elif n==999:
        return hundreds[900]+' and '+tens[90]+' ' +units[9]
    elif n==1000:
         return Specials[n] 
              
        

                

num(110)

        
```

```python
for i in range(1,1000):
    n=num(i)
    print(n)
```
